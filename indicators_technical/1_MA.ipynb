```python
import yfinance as yf
df = yf.download("SPY", start="1993-01-01")
#df
print(df.head())
```

```python
import matplotlib.pyplot as plt

# Plot the Close-Column 
plt.plot(df["Volume"])
```

```python
df["New Column"] = 0
print(df)
```

```python
# Remove New Column
df = df.drop(columns=["New Column"])
print(df)
```

```python
# Add MA Column with MA Logic (Window Size of 3)
df["MA-3"] = df["Close"].rolling(3).mean()
print(df.head())
```

```python
df["MA-50"] = df["Close"].rolling(50).mean()
df
```

```python
plt.plot(df["Close"]) #Blue
plt.plot(df["MA-50"]) #Orange
```

```python
# Calculate the moving average for a Ticker
TICKER = "SPY"
df = yf.download(TICKER, start="1993-01-01")
df["MA-50"] = df["Close"].rolling(50).mean()

plt.plot(df["Close"])
plt.plot(df["MA-50"])
plt.title(f"{TICKER} Close vs. Moving Average")
```

```python
# Calculate the moving average for a Ticker
TICKER = "SPY"
df = yf.download(TICKER, start="1993-01-01")
df["MA-50"] = df["Close"].rolling(50).mean()

#Last 100 rows & Columns, meaning the last 100 Days
df = df.iloc[-100:, :]

plt.plot(df["Close"]) # Blue
plt.plot(df["MA-50"]) #Yellow
plt.title(f"{TICKER} Close vs. Moving Average")
```
