```python
import yfinance as yf

TICKER = "SPY"
START_DATE = "1993-01-29" # SPY's first trading day
INITIAL_INVESTMENT = 1.00

# Step 1: Download data, auto-adjusting for dividends and splits.
df = yf.download(TICKER, start=START_DATE, auto_adjust=True)

# Step 2: Get the first and last adjusted prices
start_price = df['Close'].iloc[0]
end_price = df['Close'].iloc[-1]

# Step 3: Calculate the final value
final_value = INITIAL_INVESTMENT * (end_price / start_price)

print(f"An investment of €{INITIAL_INVESTMENT} on {START_DATE}")
print(f"would be worth: €{final_value} today.")
```

```python
# Method 2

# 1. Calculate the daily returns
df['daily_return'] = df['Close'].pct_change()

# 2. Calculate the cumulative growth (this is the 1€ journey)
# The .fillna(0) handles the first row being NaN
df['growth_of_1_euro'] = (1 + df['daily_return']).cumprod().fillna(1)

# 3. Get the final value
final_value = df['growth_of_1_euro'].iloc[-1]

print(f"Your €1.00 investment would be worth: €{final_value} today.")
```
